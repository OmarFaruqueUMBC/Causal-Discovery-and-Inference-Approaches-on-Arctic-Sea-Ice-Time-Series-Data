```python
# Read Data and remove the first line of variable names
X = pd.read_csv('/content/drive/My Drive/UMBC/Courses/2022_Fall/IS_800/IS 800 - group project/Data/Arctic Sea Ice/Arctic_domain_mean_1979_2021.csv',skiprows = 1, header=None)
X.head()
```

```python
model = lingam.VARLiNGAM(lags=90, criterion=None, prune=True)
model.fit(X.iloc[:,1:])
```

```python
print(model.adjacency_matrices_)
```

```python
# z: the number of time lags (0~12); i: rows of arrays; j: columns of arrays.
variables = ['wind_10m', 'specific_humidity',	'LW_down',	'SW_down',	'rainfall',	'snowfall',	'sst',	't2m',	'surface_pressure',	'sea_ice_extent']
var1_list = []
var2_list = []
lag_list = []
effect_list = []

for z in range (91):
  print('Time lag:', z)
  for i in range (10):
    for j in range (10):
      if (model.adjacency_matrices_[z,i,j] != 0) :
        var1_list.append(variables[j])
        lag_list.append(z)
        var2_list.append(variables[i])
        effect_list.append(model.adjacency_matrices_[z,i,j])
        
        print(variables[j] + '(lag=' + str(z) + ')' + ' ==> ' + variables[i] + '(lag=0)' + ', effect:' + str(model.adjacency_matrices_[z,i,j]))
  print('\n')
```

```python
# z: the number of time lags (0~12); i: rows of arrays; j: columns of arrays.
variables = ['wind_10m', 'specific_humidity',	'LW_down',	'SW_down',	'rainfall',	'snowfall',	'sst',	't2m',	'surface_pressure',	'sea_ice_extent']
var1_list = []
var2_list = []
lag_list = []
effect_list = []

for z in range (10):
  print('Effect:', variables[z])
  for i in range (91):
    for j in range (10):
      if (model.adjacency_matrices_[i,z,j] != 0) :
        var1_list.append(variables[j])
        lag_list.append(i)
        var2_list.append(variables[z])
        effect_list.append(model.adjacency_matrices_[i,z,j])
        
        print(variables[j] + '(lag=' + str(i) + ')' + ' ==> ' + variables[z] + '(lag=0)' + ', effect:' + str(model.adjacency_matrices_[i,z,j]))
  print('\n')
```

```python
all_directions = np.array(var1_list)[:, np.newaxis]
Lag = np.array(lag_list)[:, np.newaxis]
To = np.array(var2_list)[:, np.newaxis]
Effect = np.array(effect_list)[:, np.newaxis]
all_directions = np.concatenate((all_directions, Lag, To, Effect),axis=1)
all_results = pd.DataFrame(all_directions, columns = ['From','Lag','To','Effect'])
all_directions_sorted = all_results.sort_values(by=['Effect'],ascending=False)
# all_directions_sorted.to_csv('varlingam_daily_ascending.csv', index=False)
all_directions_sorted
```
